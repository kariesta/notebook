First adding this to the makefile for easy runnings:
```dumbbell-run:
	./waf --run "scratch/dumbbell-sim.cc --tcpcaa=TcpBbr --bandwidth=1Mbps"
	mkdir trace/tcpbrr
	mv *.tr trace/tcpbrr
	./waf --run "scratch/dumbbell-sim.cc --tcpcaa=TcpNewReno --bandwidth=1Mbps"
	mkdir trace/tcpnr
	mv *.tr trace/tcpnr
	./waf --run "scratch/dumbbell-sim.cc --tcpcaa=TcpCubic --bandwidth=1Mbps"
	mkdir trace/tcpcb
	mv *.tr trace/tcpcb```
 
 This runs the simulation once for each variation and results in one folder with subfolders for each TCP-variation containing all trace-files for that variation.

````python
import matplotlib.pyplot as plt

file=open("trace/rcpnr/TcpNewReno-qlen-90-bw-1Mbps-delay-50ms-cwnd.tr",'r')
lines = [line.strip().split(" ") for line in file]
linesx = [float(line[0]) for line in lines]
linesName = [line[1] for line in lines]
linesy = [float(line[2]) for line in lines]

print(lines[0])
print(linesx[0],linesy[0])
plt.plot(linesx,linesy)
plt.xlabel("miliseconds?")
plt.ylabel("cwnd size")
plt.show()
````
